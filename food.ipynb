```python
import pandas as pd 
import sqlite3
```

```python
#load csv data
orders = pd.read_csv("orders.csv")
orders.head()
```

```python
#loading json data
users = pd.read_json("users.json")
users.head()
```

```python
#showing database status
conn = sqlite3.connect("restaurants.db")
print("Database connected")
```

```python
#Searching/finding the data and loading from SQL
try:
    with open("restaurants.sql", "r", encoding="utf-8") as f:
        sql_script = f.read()

    conn.executescript(sql_script)
    print("restaurants.sql executed successfully")

except Exception as e:
    print("SQL already executed or skipped safely")
    print(e)
```

```python
#finding the restaurant data from SQL
restaurants = pd.read_sql(
    "SELECT * FROM restaurants",
    conn
)
restaurants.head()
```

```python
#displaying the restaurant columns to get more clarification
restaurants.columns
```

```python
#merging the data to get more clarity about restaurants and orders and users
orders_users = orders.merge(
    users,
    on="user_id",
    how="left"
)
final_df = orders_users.merge(
    restaurants,
    on="restaurant_id",
    how="left")
final_df = orders_users.merge(
    restaurants,
    on="restaurant_id",
    how="left")
print(final_df)
```

```python
final_df.head()
final_df.shape
final_df.to_csv("final_food_delivery_dataset.csv", index=False)
final_df.isnull().sum()
```

```python
final_df["order_date"] = pd.to_datetime(
    final_df["order_date"],
    format="%d-%m-%Y"
)
```

```python
final_df.to_csv("final_food_delivery_dataset.csv", index=False)
```
